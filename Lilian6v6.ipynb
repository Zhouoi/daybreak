GIT能够帮助我们更好地进行版本管理。 “版本控制”，也称为版本控制，源代码控制 或软件配置管理（SCM），是 **存储和管理文档、代码的更改信息**。

#### GIT常用命令和用法
+ **安装git** 参见官网安装文档 [link](https://git-scm.com/book/zh/v2/%E8%B5%B7%E6%AD%A5-%E5%AE%89%E8%A3%85-Git
)

+ **在本地创建存储库repo** *(<font color=red>请补齐操作步骤</font>)*

    1.设置git全局用户名 

    `git config --global user.name "Zhouoi" //用户名` 
    
    `git config --global user.email "464005430@qq.com" //邮箱`  

    2.初始化当前文件夹为git仓库

    `mkdir mynewproject //mkdir是用来创建一个目录`

    `cd mynewproject/ //cd是目录导航的命令`

    `git init //git初始化`

    3.设置git当前仓库用户名

    `git config user.name "DayBreak" //用户名`

    `git config user.email "464005430@qq.com" //邮箱地址`

    4.检查当前用户名设置

    `git config --list //查看git的设置情况`
    

+ **在本地创建文件、添加文件和修改文件时的git操作** *(<font color=red>请补齐操作步骤</font>)*

    1.创建本文档workshop.ipynb

    `vi workshop.ipynb //创建新的文档`

    2.添加和确认提交此文件

    `git add workshop.ipynb //添加`
    
    `git commit //提交该文件`

    3.修改文件后再次确认提交

    `git add workshop.ipynb`
  
    `git commit -m"update workshop.ipynb" //修改`

    4.查看当前仓库状态

    `git status //显示git仓库的状态`


+ **创建远程仓库[ref](https://docs.github.com/zh/get-started/importing-your-projects-to-github/importing-source-code-to-github/adding-locally-hosted-code-to-github)** *(<font color=red>请补齐操作步骤</font>)*

     1.在github上创建账号并获取ssh-key
  
     `$ ssh-keygen -t rsa -C "464005430@qq.com" //生成ssh`
     
     `cat ~/.ssh/id_rsa.pub //找到生成Key值的目录,前往.ssh目录查看对应的公钥，再在GitHub设置中添加新ssh`
     
     2.在github上创建一个没有readme、许可协议和git ignore文件的新存储库

     3.在上一步创建的本地仓库中添加远程仓库地址（github）

     `git init //初始化`

     `git remote add origin git@github.com:Zhouoi/daybreak.git //本地和远程建立联系`

     4.将本地仓库的更改推送到github中

     `git add . //暂存 `
 
     `git commit -m "first commit" //提交到本地 `

     `git push origin master //提交 `
    

+ **获取他人（组长）的远程仓库到本地并完成更改后提交** *(<font color=red>请补齐操作步骤</font>)*

    `git clone git@github.com:Zhouoi/daybreak.git //将远程仓库克隆到本地`

    `git add . //提交当前文件夹所有的文件`

    `git add [路径]`

    `git commit -m "first commit"`

    `git push origin master //将本地提交推送到远程仓库`
    

+ **组长合并他人的提交** *(<font color=red>请补齐操作步骤</font>)*

    `git checkout master`

    `git merge cate`

    `git push`
